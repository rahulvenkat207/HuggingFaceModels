```python
from transformers import pipeline
import torch
```

```python
translator = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
text = """Man for the big stage for big occasion. Bigger the game the more he wants to contribute, he likes to fight he likes to scrap as well he loves playing against Australia. Beautiful hands beautiful wrists and beautiful batsmen"""
```

```python
text_translated = translator(text,
                             src_lang="eng_Latn",
                             tgt_lang="hin_Deva")
text_translated
```
